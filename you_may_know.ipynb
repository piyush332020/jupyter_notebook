```python
import json

def load(data):
    with open(data,"r") as f:
        return json.load(f)

        
def find(user_id,data):
    user_friends={}  # id=1->key {2,3,5} ->values
    for user in data["users"]: 
        # id=1->key {2,3,5,4} ->values
        # id=2->key {1,5} ->values
        # id=3->key {5,2,1} ->values
        # id=5->key {3,2,1} ->values
        user_friends[user["id"]]=set(user['friends'])
    if user_id not in user_friends:
        return []

    direct_friends = user_friends[user_id] #{1,5}
    suggestions = {} #{3:2 , 4:1}
    for friend in direct_friends:
        for mutual in user_friends[friend]:  #id user_friend[1] => {2,3,5,4}
            if mutual!=user_id and mutual not in direct_friends: 
                suggestions[mutual]=suggestions.get(mutual,0)+1  #{3:1,4:0}
    sorted_suggestion=sorted(suggestions.items(),key= lambda x:x[1],reverse=True) #{4:0,3:1}
    return [(user_id, m) for user_id, m in sorted_suggestion]
                

data=load("massive.json")
user_id=2
rec=find(user_id,data)
print(rec)
```

```python
#pages you might like

def find_pages(user_id,data):
    #dictinory
    user_pages={}
    for user in data["users"]:
        user_pages[user['id']]=set(user["liked_pages"])
    if user_id not in user_pages:
        return []

    user_liked_pages=user_pages[user_id]
    page_suggestion={}
    for other_user,pages in user_pages.items():
        if other_user!=user_id:
            shared_pages=user_liked_pages.intersection(pages)
        for page in pages:
            if page not in user_liked_pages:
                page_suggestion[page]=page_suggestion.get(page,0)+len(shared_pages)
    sorted_page=sorted(page_suggestion.items(),key=lambda x: x[1],reverse=True)
    return [page_id for page_id,_ in sorted_page]
data=load("massive.json")
user_id=1;
recc= find_pages(user_id,data)
print(recc)
```
