```python
with open("finaldata.txt", encoding="utf-8") as f:
    data = f.read()
```

```python
chunks=data.split("\n\n")
chunks=[c for c in chunks if len(c)>3]
```

```python

print(chunks)
```

```python
def parse_chunk(chunk):
    try:
        chunk=chunk.strip()
        sep_chunk=chunk.split('\n')
        username=sep_chunk[0]
        no_of_post=int(sep_chunk[1].split(" post")[0].replace(",",""))
        
        no_of_followers=float(sep_chunk[2].split(" followers")[0].replace(",","").replace("K","").replace("M",""))
        if("K" in sep_chunk[2]):
            no_of_followers=int(no_of_followers*1000)
        elif("M" in sep_chunk[2]):
            no_of_followers=int(no_of_followers*1000000)
        else:
            no_of_followers=int(no_of_followers)

        no_of_following=float(sep_chunk[3].split(" following")[0].replace(",", "").replace("K", "").replace("M", ""))
        if("K" in sep_chunk[3]):
            no_of_following=int(no_of_following*1000)
        elif("M" in sep_chunk[3]):
            no_of_following=int(no_of_following*1000000)
        else:
            no_of_following=int(no_of_following)
        name=sep_chunk[4]  
        if(len(sep_chunk)>5):    
            type_of_page=sep_chunk[5]
            bio="\n".join(sep_chunk[6:])
        else:
            type_of_page="uknown"
            bio=""
    except Exception as e:
        print(e,chunk)
    # print(username,no_of_post,no_of_follower,no_of_following,type_of_page,bio,sep="\n")
    return {"username": username,"no_of_post":no_of_post,"no_of_followers":no_of_followers,"type_of_page":type_of_page,
            "no_of_following":no_of_following,"name":name,"bio":bio}
        
# print(parse_chunk(chunks[0]))
    
```

```python
print(parse_chunk(chunks[0]))
```

```python
all_chunks=[]
for chunk in chunks:
    parsed_chunks= parse_chunk(chunk)
    all_chunks.append(parsed_chunks)
print(all_chunks)
```
