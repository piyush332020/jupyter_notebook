```python
import pandas as pd
```

```python
emp = pd.DataFrame({
    "EmpID": [1, 2, 3],
    "Name": ["Alice", "Bob", "Charlie"],
    "DeptID": [10, 20, 30]
})

dept = pd.DataFrame({
    "DeptID": [10, 20, 40],
    "DeptName": ["HR", "Engineering", "Marketing"]
})
```

```python
pd.merge(emp,dept,on="DeptID") #inner join
```

```python
pd.merge(emp,dept,on="DeptID",how="left") #left join
```

```python
pd.merge(emp,dept,on="DeptID",how="right") #right join
```
