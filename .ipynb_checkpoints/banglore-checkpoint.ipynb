```python
def parse_chunk(chunk):
    chunk=chunk.strip()
    sep_chunk=chunk.split('\n')
    username=sep_chunk[0]
    no_of_post=sep_chunk[1]
    no_of_follower=sep_chunk[2]
    no_of_following=sep_chunk[3]
    type_of_page=sep_chunk[4]
    bio="\n".join(sep_chunk[6:])
    # print(username,no_of_post,no_of_follower,no_of_following,type_of_page,bio,sep="\n")
    return {"username": username,"no_of_post":no_of_post,"no_of_follower":no_of_follower,
            "no_of_following":no_of_following,"type_of_page":type_of_page,"bio":bio}
# print(parse_chunk(chunks[0]))
    
```
